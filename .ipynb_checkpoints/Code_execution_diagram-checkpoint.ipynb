```python
from diagrams import Cluster, Diagram, Edge
from diagrams.programming.flowchart import Action, InputOutput, ManualInput, StoredData
from diagrams.programming.language import Python

graphic_settings = {'fontsize': '20', 'bgcolor': 'transparent'}
edge_settings = {'color': 'black'}
node_settings = {}

with Diagram("Diagramme d'architecture du code", filename="Images/Code_execution_diagram", outformat='png', graph_attr=graphic_settings, edge_attr=edge_settings,
             node_attr=node_settings,
             show=False) as diag:
    
    with Cluster('Affichage des logs (post-vol)'):
        logs_display_file = Python('flight_logs_display.py')
        logs2 = InputOutput('logs.csv')
        logs_display = InputOutput('Firefox')
                                                                    
        logs2               >> Edge(xlabel='tableau csv')           >> logs_display_file
        logs_display_file   >> Edge(xlabel='Fenêtre web')           >> logs_display
    
    with Cluster('Gestion de vol en temps réel'):
        radio_communication = InputOutput('Crazyflie')
        js_input = ManualInput("Joystick")
        qtm_input = InputOutput("QTM")
        logs = InputOutput('logs.csv')

        with Cluster('Thread principal'):
            main_file = Python('main.py')
            agent_class_file = Python('agent_class.py')
            swarm_object_file = Python('swarm_object_class.py')
            qtm_tools_file = Python('qtm_tools.py')
            control_law_file = Python('uav_cotrol_law.py')

        with Cluster("Thread de gestion des entrées utilisateur"):
            js_class_file = Python('joystick_class.py')
            js_map_file = Python('joystick_map.py')
        
        js_class_file       << Edge(xlabel='Phase de vol')          << js_map_file
        js_class_file       >> Edge(xlabel='Bouton')                >> js_map_file
                                                                    
        js_input            >> Edge(xlabel='Entrée utilisateur')    >> js_class_file

        agent_class_file    << Edge(xlabel='Phase de vol du drone') << js_class_file
        agent_class_file    >> Edge(xlabel='Robot')                 >> swarm_object_file 
        agent_class_file    >> Edge(xlabel='Agent')                 >> swarm_object_file
        
        qtm_input           >> Edge(xlabel='Paquet de données')     >> main_file
        
        qtm_tools_file      >> Edge(xlabel='Position du robot')     >> agent_class_file
        qtm_tools_file      >> Edge(xlabel='Position du drone')     >> agent_class_file
        
        main_file           >> Edge(xlabel='Marqueurs')             >> qtm_tools_file
        
        swarm_object_file   << Edge(xlabel="Commande d'attitude")   << control_law_file
        swarm_object_file   >> Edge(xlabel='Robot')                 >> control_law_file
        swarm_object_file   >> Edge(xlabel='Agent')                 >> control_law_file
        
        control_law_file    >> Edge(xlabel='Données de vol')        >> logs
        
        radio_communication << Edge(xlabel="Commande d'attitude")   << swarm_object_file
    
diag
```
